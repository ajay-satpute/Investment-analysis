
# Investment Assignment


### Project Brief :
Spark Funds wants to make investments in a few companies. The CEO of Spark Funds wants to understand the global trends in investments so that she can take the investment decisions effectively.

 

### Business and Data Understanding :
Spark Funds has two minor constraints for investments:

1) It wants to invest between 5 to 15 million USD per round of investment.

2) It wants to invest only in English-speaking countries because of the ease of communication with the companies it would invest in.


### Business objective: 

The objective is to identify the best sectors, countries, and a suitable investment type for making investments. The overall strategy is to invest where others are investing, implying that the 'best' sectors and countries are the ones 'where most investors are investing'.

### Goals of data analysis:
Your goals are divided into three sub-goals:

1)Investment type analysis: Comparing the typical investment amounts in the venture, seed, angel, private equity etc. so that Spark Funds can choose the type that is best suited for their strategy.

2)Country analysis: Identifying the countries which have been the most heavily invested in the past. These will be Spark Funds’ favourites as well. 

3)Sector analysis: Understanding the distribution of investments across the eight main sectors. (Note that we are interested in the eight 'main sectors' provided in the mapping file. The two files — companies and rounds2 — have numerous sub-sector names; hence, you will need to map each sub-sector to its main sector.)

### Data cleaning :

```python
# Now finding if any columns has missing values
master_frame.isnull().any()
```

```python
# Now finding total no. of missing values per column
master_frame.isnull().sum()
```

```python
# Now finding percentage of missing values per column
round(100*(master_frame.isnull().sum()/len(master_frame.index)),2)
```

Now we are deleting following 3 columns:

1)company_permalink - As its a duplicate column because 'permalink' column contains same values

2)funding_round_code - As it has over 72% values missing and it is irrelevant column considering our data analysis goal

3)founded_at - As again its irrelevant column considering our data analysis goal

```python
# Deleting above mentioned 3 columns
master_frame = master_frame.drop('company_permalink', axis=1)
master_frame = master_frame.drop('funding_round_code', axis=1)
master_frame = master_frame.drop('founded_at', axis=1)

# Now finding percentage of missing values per column
round(100*(master_frame.isnull().sum()/len(master_frame.index)),2)
```

```python
# Deleting missing value rows from 'raised_amount_usd'

master_frame = master_frame[~np.isnan(master_frame['raised_amount_usd'])]

# Now finding percentage of missing values per column
round(100*(master_frame.isnull().sum()/len(master_frame.index)),2)
```

Now deleting columns homepage_url, state_code, region and city as these are irrelevant ones according to our data analysis goal and has some missing values.

```python
master_frame = master_frame.drop('homepage_url', axis=1)
master_frame = master_frame.drop('state_code', axis=1)
master_frame = master_frame.drop('region', axis=1)
master_frame = master_frame.drop('city', axis=1)

# Now finding percentage of missing values per column

round(100*(master_frame.isnull().sum()/len(master_frame.index)),2)
```

Now deleting missing value rows from category_list and country_code column as these are useless for our analysis.

```python
master_frame=master_frame[~(master_frame['category_list'].isnull())]
master_frame=master_frame[~(master_frame['country_code'].isnull())]
```

```python
# Now finding percentage of missing values per column

round(100*(master_frame.isnull().sum()/len(master_frame.index)),2)
```

### 1) Funding type analysis :

```python
# Now Spark Funds wants to choose one of the four investment types which are seed, angel, venture and private equity
# for each potential investment they will make. So we will now consider the rows which has only this funding_round_type.

master_frame = master_frame[(master_frame.funding_round_type == 'seed')|(master_frame.funding_round_type == 'angel')|(master_frame.funding_round_type == 'venture')|(master_frame.funding_round_type == 'private_equity')]

master_frame.head()
```

### 3) Sector analysis 1 :

1) Extract the primary sector of each category list from the category_list column

2) Use the mapping file 'mapping.csv' to map each primary sector to one of the eight main sectors (Note that ‘Others’ is also considered one of the main sectors)

```python
master_frame=master_frame[(master_frame.country_code=='USA')|(master_frame.country_code=='GBR')|(master_frame.country_code=='IND')]
master_frame.head()
```

Now extracting the primary sector of each category list from the category_list column and making new **primary_sector** column.

```python
# Defining a function prim_sector so that we can use it to create a new column in our master_frame

def prim_sector(x):
        x=x.split('|')[0]
        return(x)
    
# Applying function prim_sector

master_frame['primary_sector']=master_frame['category_list'].apply(prim_sector)
master_frame.head()
```

```python
# Reading mapping.csv file
mapping=pd.read_csv("mapping.csv",encoding = "ISO-8859-1")
mapping.head()
```

```python
mapping= pd.melt(mapping,id_vars=['category_list'],value_vars=['Automotive & Sports','Cleantech / Semiconductors','Entertainment','Health','Manufacturing','News, Search and Messaging','Others','Social, Finance, Analytics, Advertising'])

# In mapping category_list 'na is misspellt as 0', so we are going to change that. 
mapping.category_list = mapping.category_list.replace({'0':'na',}, regex=True)

# Now lets see mapping dataframe
mapping.head()
```

```python
mapping = mapping[~(mapping.value == 0)]
mapping = mapping.drop('value', axis = 1)
mapping = mapping.rename(columns = {'category_list':'primary_sector','variable':'main_sector'})
mapping.head()
```

```python
mapping.shape
```

```python
mapping.info()
```

Now we are going to merge these 2 dataframes with each primary sector mapped to its main sector.

```python
master_frame = master_frame.merge(mapping, how = 'left', on ='primary_sector')
master_frame.head()
```

```python
master_frame.main_sector.isnull().sum()
```

So we are going to delete the rows which doesn't have main sector values

```python
master_frame=master_frame[~(master_frame.main_sector.isnull())]
master_frame.head()
```

```python
master_frame.info()
```

### 4) Sector analysis 2 :

Create three separate data frames D1, D2 and D3 for each of the three countries containing the observations of funding type FT falling within the 5-15 million USD range. The three data frames should contain:

1) All the columns of the master_frame along with the primary sector and the main sector.

2) The total number (or count) of investments for each main sector in a separate column.

3) The total amount invested in each main sector in a separate column.

```python
# Creating D1 dataframe for USA with FT as 'venture' falling within the 5-15 million USD range. 

D1= master_frame[(master_frame.country_code=='USA')&(master_frame.raised_amount_usd>=5)&(master_frame.raised_amount_usd<=15)]

D1.head()
```

```python
D1_n=D1.pivot_table(values='raised_amount_usd',index='main_sector',aggfunc=['sum','count'])
D1_n.head(8)
```

```python
D1_n=D1_n.rename(columns={'sum':'Total_amount','count':'Total_count'})
D1_n.head(8)
```

```python
# Now merging these 2 dataframes
D1=D1.merge(D1_n,how='left',on='main_sector')
D1.head()
```

```python
# Now similarly creating D2 dataframe for 'GBR' with FT as 'venture' falling within the 5-15 million USD range. 
D2= master_frame[(master_frame.country_code=='GBR')&(master_frame.raised_amount_usd>=5)&(master_frame.raised_amount_usd<=15)]
D2_n=D2.pivot_table(values='raised_amount_usd',index='main_sector',aggfunc=['sum','count'])
D2_n=D2_n.rename(columns={'sum':'Total_amount','count':'Total_count'})
D2=D2.merge(D2_n,how='left',on='main_sector')
D2.head()
```

```python
# Now similarly creating D3 dataframe for 'IND' with FT as 'venture' falling within the 5-15 million USD range. 
D3= master_frame[(master_frame.country_code=='IND')&(master_frame.raised_amount_usd>=5)&(master_frame.raised_amount_usd<=15)]
D3_n=D3.pivot_table(values='raised_amount_usd',index='main_sector',aggfunc=['sum','count'])
D3_n=D3_n.rename(columns={'sum':'Total_amount','count':'Total_count'})
D3=D3.merge(D3_n,how='left',on='main_sector')
D3.head()
```

```python
# Now finding out total number and amount of investment for 3 countries
print('Total number of investments (count) for 3 countries is as follows:')
print('USA=', D1.raised_amount_usd.count())
print('GBR=', D2.raised_amount_usd.count())
print('IND=', D3.raised_amount_usd.count())
print('\n')
print('Total amount of investment (USD) for 3 countries is as follows:')
print('USA=', round(D1.raised_amount_usd.sum(),2))
print('GBR=', round(D2.raised_amount_usd.sum(),2))
print('IND=', round(D3.raised_amount_usd.sum(),2))
```
